## Linear Regression

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
````

````python
data= pd.read_csv("./DataSets/kc_house_data.csv")
data.drop(columns=["id"], inplace=True)
data.head()
````

````python
plt.figure(figsize=(30,18))
corr= data.corr()
upper= np.triu(corr)
sns.heatmap(corr, mask=upper, annot=True, )
````

````python
corr.head()
````

````python
values= corr.iloc[0].copy()
# values.drop("price", inplace=True)
````

````python
featues_selected= values[values>=0.25].index
````

````python
new_data= data[featues_selected]
new_data.head()
````

````python
sns.heatmap(data.isnull())
````

````python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV, ElasticNet, ElasticNetCV
````

````python
x=new_data.iloc[:,1:]
y= new_data["price"]
````

````python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.4, random_state=42)
````

####  with Linear Regression

````python
# %config Completer.use_jedi = False
````

````python
model = LinearRegression(fit_intercept=True)

pipeline= make_pipeline(StandardScaler(), model)

pipeline.fit(x_train, y_train)

preds= pipeline.predict(x_test)

r2= r2_score(y_test, preds)
mse= mean_squared_error(y_test, preds)

print(r2, mse)

# 65 -> r2 score
````

#### with Ridge Regression

- [link](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html)

````python
model = Ridge(alpha=2.0,fit_intercept=True)

pipeline= make_pipeline(StandardScaler(), model)

pipeline.fit(x_train, y_train)

preds= pipeline.predict(x_test)

r2= r2_score(y_test, preds)
mse= mean_squared_error(y_test, preds)

print(r2, mse)

# 65 -> r2 score
````

#### with RidgeCV

- [link](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RidgeCV.html#sklearn.linear_model.RidgeCV)

````python
model = RidgeCV(alphas= np.linspace(1,100,200) , fit_intercept=True, cv=10, scoring='neg_mean_squared_error')

pipeline= make_pipeline(StandardScaler(), model)

pipeline.fit(x_train, y_train)

preds= pipeline.predict(x_test)

r2= r2_score(y_test, preds)
mse= mean_squared_error(y_test, preds)

print(r2, mse)

# 65 -> r2 score
````

#### with Lasso Regression

- [link](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html#sklearn.linear_model.Lasso)

````python
# https://stackoverflow.com/a/20682033/12210002
model = Lasso(alpha=2.0,fit_intercept=True, tol=0.05)

pipeline= make_pipeline(StandardScaler(), model)

pipeline.fit(x_train, y_train)

preds= pipeline.predict(x_test)

r2= r2_score(y_test, preds)
mse= mean_squared_error(y_test, preds)

print(r2, mse)

# 65 -> r2 score
````

#### with LassoCV

- [link](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoCV.html#sklearn.linear_model.LassoCV)

````python
model = LassoCV(alphas= np.linspace(1,200,300) , fit_intercept=True, cv=15)

pipeline= make_pipeline(StandardScaler(), model)

pipeline.fit(x_train, y_train)

preds= pipeline.predict(x_test)

r2= r2_score(y_test, preds)
mse= mean_squared_error(y_test, preds)

print(r2, mse)

# 65 -> r2 score
````

#### using Elasticnet

- [link](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.ElasticNet.html#sklearn.linear_model.ElasticNet)

```
1 / (2 * n_samples) * ||y - Xw||^2_2
+ alpha * l1_ratio * ||w||_1
+ 0.5 * alpha * (1 - l1_ratio) * ||w||^2_2
```

````python

model = ElasticNet(alpha=1.0, l1_ratio=0.99,fit_intercept=True)

pipeline= make_pipeline(StandardScaler(), model)

pipeline.fit(x_train, y_train)

preds= pipeline.predict(x_test)

r2= r2_score(y_test, preds)
mse= mean_squared_error(y_test, preds)

print(r2, mse)

# 65 -> r2 score
````

#### using ElasticnetCV

- [link](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.ElasticNetCV.html#sklearn.linear_model.ElasticNetCV)

````python
model = ElasticNetCV(l1_ratio= np.linspace(0.1,1,10), alphas= np.linspace(1,200,300) , fit_intercept=True, cv=15, tol=0.1)

pipeline= make_pipeline(StandardScaler(), model)

pipeline.fit(x_train, y_train)

preds= pipeline.predict(x_test)

r2= r2_score(y_test, preds)
mse= mean_squared_error(y_test, preds)

print(r2, mse)

# 65 -> r2 score
````
