```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data= pd.read_csv("./DataSets/FuelEconomy.csv")
data.head()
```

```python
# Renaming columns
data.rename(columns={"Horse Power" : "HP", "Fuel Economy (MPG)": "MPG"}, inplace= True)
data.tail()
```

#### Preparing for training and testing dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x= data[["HP"]]
y= data["MPG"]
```

```python
# Splitting the data : for training and testing
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.4, random_state= 42)
```

```python
# Training the data with Linear Regression model
model= LinearRegression(fit_intercept=True)

model.fit(x_train, y_train)
```

```python
# Getting the coefficients and intercepts
model.coef_, model.intercept_
```

```python
# Predciting for test set
preds= model.predict(x_test)
```

```python
# Checking the results
r2= r2_score(y_test, preds)
mse= mean_squared_error(y_test, preds)

r2, mse
```

```python
# Plotting
plt.scatter(x_test, y_test)
plt.plot(x_test, preds, color='r')
plt.title("Linear Regression 1")
plt.xlabel("HP")
plt.ylabel("MPG")
plt.plot()
```

```python
sns.kdeplot(y_test-preds)
```

```python
# Predcting for a given value

inputs= [[10]] # Single Feature

model.predict(inputs)[0]
```
